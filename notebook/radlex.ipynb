```python
import pandas as pd
import os

def extract_columns(input_file, output_file):
    # Determine the file extension
    _, file_extension = os.path.splitext(input_file)
    
    # Read the Excel file
    if file_extension.lower() == '.xls':
        # For .xls files, we need to use the 'xlrd' engine
        df = pd.read_excel(input_file, engine='xlrd')
    else:
        # For .xlsx files, we can use the default engine
        df = pd.read_excel(input_file)

    # List of columns to extract
    columns_to_extract = [
        'Class ID',
        'Preferred Label',
        'Preferred_name',
        'Synonyms',
        'Synonym',
        'Definitions',
        'Definition',
        'Parents',
        'Is_A',
        'Continuous_With',
        'FMAID',
        'SNOMED_ID',
        'UMLS_ID',
        'Neurolex_ID',
        'http://data.bioontology.org/metadata/prefixIRI'
    ]

    # Extract only the specified columns that exist in the dataframe
    existing_columns = [col for col in columns_to_extract if col in df.columns]
    extracted_df = df[existing_columns]

    # Save the extracted data to a new Excel file
    extracted_df.to_excel(output_file, index=False, engine='openpyxl')

    print(f"Extracted data saved to {output_file}")

# Example usage
input_file = '/home/hbchen/Projects/MedKGC/resource/RadLex Term Browser.xls'
output_file = '/home/hbchen/Projects/MedKGC/resource/redlexX.xls'

extract_columns(input_file, output_file)

import pandas as pd

# Read the Excel file
df = pd.read_excel('/home/hbchen/Projects/MedKGC/resource/redlexX.xls')

# Print the contents of the file
df
```

### morphological search unsing levenshtein distance

```python
import pandas as pd
import numpy as np
from Levenshtein import ratio

def safe_levenshtein_ratio(query, x):
    """
    安全地计算Levenshtein比率，处理非字符串输入。
    """
    if pd.isna(x) or not isinstance(x, str):
        return 0.0  # 对于非字符串或NaN值，返回0相似度
    try:
        return ratio(str(query), str(x))
    except Exception:
        return 0.0  # 如果计算失败，返回0相似度

def find_similar_entities(query, df, column="Preferred Label", top_n=5):
    """
    查找与输入查询最相似的实体。
    
    参数:
    query (str): 要查找的实体名称
    df (pandas.DataFrame): 包含实体名称的DataFrame
    column (str): 包含实体名称的列名
    top_n (int): 返回的最相似实体数量
    
    返回:
    list of tuples: 包含(实体名称, 相似度分数)的列表,按相似度降序排序
    """
    # 确保查询是字符串
    query = str(query)
    
    # 计算查询与每个实体名称的Levenshtein比率
    similarities = df[column].apply(lambda x: safe_levenshtein_ratio(query, x))
    
    # 创建包含实体名称和相似度分数的DataFrame
    result_df = pd.DataFrame({
        'entity': df[column],
        'similarity': similarities
    })
    
    # 按相似度降序排序并返回top_n个结果
    return result_df.sort_values('similarity', ascending=False).head(top_n).values.tolist()

# 使用示例
# 假设df是您的DataFrame,包含"Preferred Label"列
query = "lung"

try:
    similar_entities = find_similar_entities(query, df)
    print(f"与'{query}'最相似的实体:")
    for entity, similarity in similar_entities:
        print(f"{entity}: {similarity:.2f}")
except Exception as e:
    print(f"发生错误: {e}")
    print("DataFrame的前几行:")
    print(df[["Preferred Label"]].head().to_string())
    print("\nPreferred Label列的数据类型:")
    print(df["Preferred Label"].dtype)
    print("\n非字符串值的示例 (如果有):")
    non_string = df[df["Preferred Label"].apply(lambda x: not isinstance(x, str))]
    print(non_string["Preferred Label"].head().to_string())
```

### semantic search

#### word_centric_embedding

```python
# 定义项目
items = [
    "Pneumothorax is An accumulation of air or gas in the PLEURAL CAVITY, which may occur spontaneously or as a result of trauma or a pathological process. The gas may also be introduced deliberately during PNEUMOTHORAX, ARTIFICIAL. (MSH)",
    "Lungs is Either of the pair of organs occupying the cavity of the thorax that effect the aeration of the blood. (MSH)",
    "Pleural is Of or pertaining to the pleura. (NCI)",
    "Thyroidectomy is Surgical removal of the thyroid gland."
]

mention = "Pneumothorax"
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from scipy.spatial.distance import euclidean
from transformers import AutoTokenizer, AutoModel
import torch

# 首先，我们需要重新定义get_word_centric_embedding函数
def get_word_centric_embedding(text, target_word, model_name="bert-base-uncased"):
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModel.from_pretrained(model_name)

    inputs = tokenizer(text, return_tensors="pt", add_special_tokens=True)
    
    target_word_ids = tokenizer.encode(target_word, add_special_tokens=False)
    
    target_positions = []
    for i in range(len(inputs['input_ids'][0]) - len(target_word_ids) + 1):
        if inputs['input_ids'][0][i:i+len(target_word_ids)].tolist() == target_word_ids:
            target_positions.extend(range(i, i+len(target_word_ids)))

    with torch.no_grad():
        outputs = model(**inputs)

    word_embeddings = outputs.last_hidden_state[0, target_positions, :]
    word_centric_embedding = torch.mean(word_embeddings, dim=0)

    return word_centric_embedding.numpy()

# 定义相似性计算函数
def cosine_sim(a, b):
    return cosine_similarity([a], [b])[0][0]

def euclidean_sim(a, b):
    return 1 / (1 + euclidean(a, b))

def dot_product_sim(a, b):
    return np.dot(a, b)

# 获取每个项目的target word和embedding
item_embeddings = []
for item in items:
    target_word = item.split(' is')[0]
    embedding = get_word_centric_embedding(item, target_word)
    item_embeddings.append(embedding)

# 获取mention的embedding
mention_embedding = get_word_centric_embedding(mention, mention)


# 计算相似性并排序
similarity_methods = {
    "Cosine Similarity": cosine_sim,
    "Euclidean Similarity": euclidean_sim,
    "Dot Product": dot_product_sim
}

for method_name, sim_func in similarity_methods.items():
    print(f"\nUsing {method_name}:")
    similarities = [sim_func(mention_embedding, item_emb) for item_emb in item_embeddings]
    sorted_indices = np.argsort(similarities)[::-1]
    
    for i, idx in enumerate(sorted_indices):
        print(f"{i+1}. {items[idx][:50]}... (Similarity: {similarities[idx]:.4f})")
```

这些结果很有趣，让我们来分析一下这三种相似度计算方法的表现：

1. 余弦相似度 (Cosine Similarity)：

余弦相似度测量两个向量之间的角度，范围通常在-1到1之间，1表示方向完全相同，0表示正交，-1表示方向相反。在我们的结果中，值域在0到1之间，这是因为我们使用的是非负的词嵌入。

优点：
- 对向量的大小不敏感，主要关注方向。
- 结果易于解释，0.6856的相似度表示相当高的相似性。
- 清晰地区分了最相关的项（Pneumothorax）和其他项。

观察：
- Pneumothorax和Pleural之间的相似度差距明显，这符合预期，因为Pleural（胸膜）与Pneumothorax（气胸）在医学上确实相关。
- Lungs（肺）的相似度最低，这可能是因为虽然肺与气胸有关，但在语义表示上差异较大。

2. 欧几里得相似度 (Euclidean Similarity)：

这里使用的是欧几里得距离的倒数，因此较小的距离会产生较大的相似度值。

优点：
- 考虑了向量在所有维度上的绝对差异。
- 对数值的绝对大小敏感。

观察：
- 总体趋势与余弦相似度一致，但数值范围较小（0.0546到0.1000）。
- 区分度似乎不如余弦相似度明显，特别是在较低相似度的项目之间。

3. 点积 (Dot Product)：

点积直接计算两个向量对应元素的乘积和，没有归一化。

优点：
- 计算简单快速。
- 保留了向量大小的信息。

观察：
- 结果的数值范围较大（32.2225到86.9239）。
- 排序结果与其他两种方法一致，表明它在这个场景下也能有效区分相似度。
- 但由于缺乏归一化，可能难以跨不同长度的文本或不同规模的嵌入进行比较。

总体分析：

1. 一致性：所有三种方法都正确地识别出Pneumothorax定义为最相似的项，这证实了我们的方法在捕捉语义相似性方面的有效性。

2. 排序一致性：三种方法给出的排序完全一致，这增强了我们对结果可靠性的信心。

3. 区分能力：
   - 余弦相似度在区分高相似度和低相似度项目方面表现最好。
   - 欧几里得相似度的数值范围较小，可能需要更精细的阈值来区分相似度级别。
   - 点积提供了最大的数值范围，但缺乏标准化可能使其难以在不同场景中设置统一的阈值。

4. 语义捕捉：结果显示Pleural（胸膜）比Lungs（肺）更相似于Pneumothorax（气胸），这在医学上是合理的，因为气胸直接涉及胸膜腔。

5. 应用建议：
   - 对于需要标准化比较的场景，余弦相似度可能是最佳选择。
   - 如果向量的大小信息很重要，点积可能更合适。
   - 欧几里得相似度在这个场景中表现不如其他两种方法明显，但在某些特定应用中可能有其优势。

6. 改进空间：考虑到Lungs与Pneumothorax的相关性在医学上应该较高，但在我们的结果中相似度最低，这可能表明我们的嵌入方法或相似度计算在捕捉某些语义关系方面还有改进空间。

这个分析为我们提供了有价值的见解，可以指导我们在实际应用中选择合适的相似度度量方法，并帮助我们理解如何进一步优化我们的语义表示和相似度计算方法。
